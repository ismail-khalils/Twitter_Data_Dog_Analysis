### Issue #8: None in doggo, floofer, pupper, puppo columns are treated as non-null values

#### Define I will remove the None values with np.nan

#### Code

```python
merged_df.replace({'doggo': {'None': np.nan}, 
                    'floofer': {'None': np.nan}, 
                    'pupper': {'None': np.nan}, 
                    'puppo': {'None': np.nan}}, inplace=True)
```

#### Test

```python
merged_df.query('doggo == "None" or floofer == "None" or pupper == "None" or puppo == "None"').count()
```

### Issue #9: Combining the dog stage columns 

####  Replacing the none in stage columns with empty strings, combining them, and formatting the entries with mutiple dog stages

#### Code

```python
# First replace None in stage columns with empty string as follows.
merged_df.doggo.replace('None', '', inplace=True)  # Like this do replacements for all four stages
merged_df.pupper.replace('None', '', inplace=True)  
merged_df.puppo.replace('None', '', inplace=True)  
merged_df.floofer.replace('None', '', inplace=True)  

# Then combine stage columns.
merged_df['stage'] = merged_df.doggo + merged_df.floofer + merged_df.pupper + merged_df.puppo

# Then format entries with multiple dog stages which appeared like doggopupper.

merged_df.loc[merged_df.stage == 'doggopupper', 'stage'] = 'doggo, pupper' 
merged_df.loc[merged_df.stage == 'doggopuppo', 'stage'] = 'doggo, puppo' 
merged_df.loc[merged_df.stage == 'doggofloofer', 'stage'] = 'doggo, floofer'
```

#### Test

```python
merged_df
```

####  Merging everything together

```python
Required_copy['tweet_id'] = Required_copy['tweet_id'].astype(str)
merged_df2 = pd.merge(merged_df, Required_copy, on='tweet_id')
```
